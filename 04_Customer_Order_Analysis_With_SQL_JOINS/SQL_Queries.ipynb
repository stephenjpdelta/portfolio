```python
# ==========================================
# SECTION 1: Imports + Connect to Database
# ==========================================

import sqlite3
import pandas as pd

conn = sqlite3.connect("CustomerOrders.db")
```

```python
# ==========================================
# SECTION 2: Query 1 - Inner Join
# ==========================================

query = """
SELECT
    c.name,
    o.order_id,
    o.order_date,
    o.quantity
FROM customers c
INNER JOIN orders o
    ON c.customer_id = o.customer_id;
"""

df = pd.read_sql_query(query, conn)
print(df)
```

```python
# ==========================================
# SECTION 3: Query 2 - Left Join
# ==========================================

query = """
SELECT
    c.name,
    o.order_id,
    o.order_date
FROM customers c
LEFT JOIN orders o
    ON c.customer_id = o.customer_id;
"""

df = pd.read_sql_query(query, conn)
print(df)
```

```python
# ==========================================
# SECTION 4: Query 3 - Total Spent by Customer
# ==========================================

query = """
SELECT
    c.name,
    SUM(o.quantity * p.price) AS total_spent
FROM customers c
LEFT JOIN orders o
    ON c.customer_id = o.customer_id
LEFT JOIN products p
    ON o.product_id = p.product_id
GROUP BY c.customer_id, c.name;
"""

df = pd.read_sql_query(query, conn)
print(df)
```

```python
# ==========================================
# SECTION 5: Query 4 - Customers With No Orders
# ==========================================

query = """
SELECT
    c.customer_id,
    c.name
FROM customers c
LEFT JOIN orders o
    ON c.customer_id = o.customer_id
WHERE o.order_id IS NULL;
"""

df = pd.read_sql_query(query, conn)
print(df)
```

```python
# ==========================================
# SECTION 6: Query 5 - Orders with Product Details
# ==========================================

query = """
SELECT
    c.name AS customer_name,
    p.product_name,
    p.category,
    o.quantity,
    o.order_date
FROM orders o
INNER JOIN customers c
    ON o.customer_id = c.customer_id
INNER JOIN products p
    ON o.product_id = p.product_id;
"""

df = pd.read_sql_query(query, conn)
print(df)
```

```python
# ==========================================
# SECTION 7: Query 6 - Revenue by Product
# ==========================================

query = """
SELECT
    p.product_name,
    p.category,
    SUM(o.quantity) AS total_units_sold,
    SUM(o.quantity * p.price) AS total_revenue
FROM orders o
INNER JOIN products p
    ON o.product_id = p.product_id
GROUP BY p.product_id, p.product_name, p.category;
"""

df = pd.read_sql_query(query, conn)
print(df)
```

```python
# ==========================================
# SECTION 8: Close Connection
# ==========================================

conn.close()
```
