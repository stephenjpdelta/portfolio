# Linear Regression Analysis using Scikit-Learn

This project demonstrates how to build a linear regression model using Python and scikit-learn.

## Objective
Predict student scores based on study hours.

## Tools Used
- Python
- pandas
- numpy
- scikit-learn
- matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

## Load Dataset

```python
data = pd.read_csv("C:\\Users\\steve\\Documents\\Datasets\\Raw\\boston_house_prices\\Boston-house-price-data.csv")
print(data.head())
```

## Data Exploration

```python
print(data.describe())
```

## Define Features and Target

```python
x = data[['RM']]
y = data[['TAX']]
```

## Split the Dataset

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=41
)
```

## Train the Model

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

## Make Predictions

```python
y_pred = model.predict(x_test)
```

## Model Evaluation

```python
print("Mean Absolute Error:", round(metrics.mean_absolute_error(y_test, y_pred), 2))
print("Mean Squared Error:", round(metrics.mean_squared_error(y_test, y_pred), 2))
print("R2 Score:", round(metrics.r2_score(y_test, y_pred), 2))
```

## Visualisation

```python
plt.scatter(x_test, y_test, color='blue')
plt.plot(x_test, y_pred, color='red')
plt.xlabel('Average Rooms per Dwelling')
plt.ylabel('Full-value Property Tax Rate per $10,000')
plt.title('Linear Regression')
plt.show()
```

## Predict New Values

```python
new_value = pd.DataFrame({'RM': [6]})
prediction = model.predict(new_value)
print("Predicted Score:", round(prediction[0].item(), 2))
```
